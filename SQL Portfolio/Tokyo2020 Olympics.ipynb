**QUESTION 1:** 
Number of athletes in the competition

```python
%%sql
SELECT COUNT(DISTINCT NAME) FROM athlete
```

**QUESTION 2**: Number of participating country

```python
%%sql
SELECT COUNT(DISTINCT NOC) FROM athlete
```

**QUESTION 3**: Countries with the highest number of athletes


```python
%%sql

SELECT NOC, COUNT(DISTINCT Name) as total_athletes from athlete
Group by 1
order by 2 Desc 
Limit 10
```

**QUESTION4**: Top 10 events( by number of participation)

```python
%%sql

SELECT Discipline, COUNT(DISTINCT Name) as total_athletes from athlete
Group by 1
order by 2 desc 
Limit 10
```

**QUESTION 5**: Bottom 10 events (by participation)

```python
%%sql
SELECT Discipline, COUNT(DISTINCT Name) as total_athletes from athlete
Group by 1
order by 2 asc 
Limit 10
```

**Question 6**: Percentage of medals won

---



```python
%%sql
SELECT "Team/NOC",Total,(Total*100.0/(SELECT SUM(Total) From medal))  'pct(%)'
From medal
Group by 1
order by 3 desc


```

**QUESTION 7**: what sporting activities did Nigeria participate in

```python
%%sql

SELECT Name ,Discipline,Event
FROM team
where Name ="Nigeria"
```

QUESTION 8: How many medals did Nigeria win

```python
%%sql

SELECT "Team/NOC", Gold, Silver,Bronze,Total
FROM medal

where "Team/NOC"="Nigeria"
```

**QUESTION 9**: Identify athletes who partook in the 2020 Olympics for Nigeria

```python
%%sql
SELECT "Team/NOC", team.Event,team.Discipline, athlete.Name
FROM medal
JOIN team on medal."Team/NOC" = team.Name
JOIN athlete on medal."Team/NOC"= athlete.NOC
where "Team/NOC"="Nigeria"
```

**QUESTION 10**: *Athletes who partook in Basketball*

```python
%%sql
SELECT ( "Team/NOC" ),team.Event,team.Discipline, athlete.Name
FROM medal
JOIN team on medal."Team/NOC" = team.Name
JOIN athlete on medal."Team/NOC"= athlete.NOC
where "Team/NOC"="Nigeria" and team.Discipline ="Basketball"
```
